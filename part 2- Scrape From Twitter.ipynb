## Data Cleaning

Then we will copy and paste this data into Excel and conduct the initial data cleaning. Then we will use python to check the duplicate & null values.

```python
#read tables
userinfo=pd.read_csv("Final_Data/twitterUser.csv",encoding = 'latin-1')
taginfo=pd.read_csv("Final_Data/twitterTag.csv",encoding='latin-1')
```

```python
# check null value
userinfo.isnull().sum()
```

```python
# remove null value
userinfo_clean=userinfo.dropna()
```

```python
#recheck if it is working
userinfo_clean.isnull().sum()
```

```python
#check duplicate values
userinfo_clean.duplicated().sum()
```

```python
# remove duplicate values
twitterUser = userinfo_clean.drop_duplicates(keep = 'first')
```

```python
# recheck duplicate values
twitterUser.duplicated().sum()
```

```python
# output the final tables
twitterUser.to_csv('twitterUser.csv')
```
