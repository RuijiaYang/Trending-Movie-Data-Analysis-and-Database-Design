# Part 7- Sentiment Analysis

Sentiment Analysis here is to find the categories of tags.We will be using WordNet hypernyms function to conduct it.
Hypernyms means a bigger concept here. And in a way can show the meaning of this tag.

The following example here show that the hypernyms of "woman" will be adult or female, which to a certain extent show it's sentiment.

```python
import pandas as pd
from nltk.corpus import wordnet as wn
```

```python
#import tag tables
twitterTag = pd.read_csv("Final_Data/twitterTag.csv",encoding = 'latin-1')
youtubeData = pd.read_csv("Final_Data/youtubeData_clean.csv",encoding = 'latin-1')
```

```python
# show the lenth of twittertag
tag=twitterTag['tag']
print(len(tag))
```

```python
# print all the existing hypernyms of tags in the form of csv
for i in range(1,2524):  
    #syns_list=[]
    syn = wn.synsets(tag[i])
    if(syn!=[]):
        syn = wn.synsets(tag[i])[0]
        syns=syn.hypernyms()
#        if(syns!=[]):
        print(i,end=',')
        print(tag[i],end=',')
        for j in syns:
            print(j.lemma_names(),end=',')
```

Then we copy those data into csv tables and conduct the initial data cleaning.

```python
# import the roughly cleaned datasets and remove duplicates and fill null values.
sentiment=pd.read_csv('Final_Data/sentiment1.CSV')
sentiment=sentiment.fillna('none')
sentiment=sentiment.drop_duplicates()
sentiment.index=range(len(sentiment))
sentiment[0:1000:100]
```

```python
# create a function to remove irrelevant symbols 
def normalization(col):
    for r in range(0,359):
        if (col[r]!='none'):
            col[r]=str(col[r])
            location=col[r].find("'")
            if(location!=-1):
                 col[r]=col[r][0:location]
```

```python
# apply function over 3 columns
normalization(sentiment['sentiment_1'])
normalization(sentiment['sentiment_2'])
normalization(sentiment['sentiment_3'])
```
